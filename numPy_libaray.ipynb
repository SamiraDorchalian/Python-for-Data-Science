## vstack hstack concantenate

```python
day1= np.array([[1,2],[3,4]])
day2= np.array([[11,2],[6,5]])
np.vstack((day1, day2))
```

```python
np.hstack((day1, day2))
```

```python
np.concatenate((day1, day2), axis=0)
```

```python
day1= np.array([[1,2],[3,4]])
day2= np.array([[11,2],[6,5]])
np.concatenate((day1, day2), axis=1)
```
